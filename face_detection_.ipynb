```python
import cv2
import matplotlib.pyplot as plt
```

```python
#face_classifier=cv2.CascadeClassifier('Haarcascades/haarcascade_frontalface_default.xml')
face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
print(face_classifier)              
```

```python
img=cv2.imread("ais.jpeg")
# gray_2=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
cv2.imshow("img",img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
faces = face_classifier.detectMultiScale(img,1.3,5)
# if faces is not 
if faces is ():
    print("No faces found")

print(faces)
```

```python
# Start coordinate, here (5, 5) (x,y)
# represents the top left corner of rectangle 
# Ending coordinate, here (220, 220) (x+w,y+h)
# represents the bottom right corner of rectangle 
# Blue color in BGR (212,4,54)
# Line thickness of 2 px 
for (x,y,w,h) in faces:
    cv2.rectangle(img,(x,y),(x+w,y+h),(212,4,54),2)
    cv2.imshow('face_detection',img)
    cv2.cv2.waitKey()
cv2.destroyAllWindows()
plt.imshow(img)
plt.show()
```
